### method overloading
- allow programmers to write methods that can process a variety of different types of functinalities with same name
- Method overloading is a compile type polymorphism using which we can define two or more methods in same class with the same name but different with parameter

### method overriding
- A child class method override parent class method of the same name, parameter, and return type.
- it is used to overwrite or redefine the parent class method in derived class

### operator overloading
- giving extended meaning beyond their predefined operational meaning.

### Mathematical operators
- + : __add__(self,other)
- - : __sub__(self,other)
- * : __mul__(self,other)
- / : __truediv__(self,other)
- // : __floordiv__(self,other)
- % : __mod__(self,other)
- ** : __pow__(self,other)

### comparison operator
- > : __gt__(self,other)
- < : __lt__(self,other)
- >= : __ge__(self,other)
- <= : __le__(self,other)
- == : __eq__(self,other)
- != : __ne__(self,other)

### Inheritance
- we can use existing class to create a new class rather than creating it from scratch
- allow the program to reuse the code
- with inheritance, child class gain access to all data members, function and properties defined in the parent class
- A child may also offer it's particular implementation of the parent class function
- a derived class can be inherit from its base class b simply putting its name in the paranthesis afhter the derived class name.
- different 5 types of inheritance
- 1) single inheritance(simple inheritance)
- 2) multiple inheritance
- 3) multi level inheritance
- 4) hierarchial inheritance
- 5) hybrid inheritance
- concept of MRO
-  Abstract Class

```python
def sum_num(*args):
    result=0
    for num in args:
        result+=num
    print('sum: ',result)
print("similar to method overloading concept")
print("single argument ->",end=" ")
sum_num(10)
print("double argument ->",end=" ")
sum_num(10,20)
print("penta argument ->",end=" ")
sum_num(10,20,30,40,50)
```

```python
class A:
    def first(self):
        print("First function of class A")
    def second(self):
        print("second function of class A")
class B(A):
    def first(self):           #overriden method
        print("First function of class B")
    def display(self):
        print("display function of class B")
```

```python
obj=B()
obj.first()
obj.second()
obj.display()
```

```python
class A:
    def __init__(self,a):
        self.a=a
    def __add__(self,other):
        return self.a+other.a
    def __gt__(self,other):
        if self.a>other.a:
            return True
        else:
            return False
        
obj1=A(5)
obj2=A(8)
print(obj1+obj2)
# or print(obj1.__add__(obj2))
# or print(A.__add__(obj1,obj2))
obj3=A("hello")
obj4=A("hiii")
print(obj3+obj4)
# or print(obj3.__add__(obj4))
# or print(A.__add__(obj3,obj4))
if obj1>obj2:
    print("obj1 is greater")
else:
    print("obj2 is greater")
if obj3>obj4:
    print("obj3 is greater")
else:
    print("obj4 is greater")
```

```python
class C:
    def __init__(self,a):
        self.a=a
    def __sub__(self,other):
        return self.a-other.a
    def __mul__(self,other):
        return self.a*other.a
    def __truediv__(self,other):
        return self.a/other.a
    def __floordiv__(self,other):
        return self.a//other.a
    def __mod__(self,other):
        return self.a%other.a
    def __pow__(self,other):
        return self.a**other.a
obj1=C(4)
obj2=C(2)
print(obj1-obj2)
print(obj1*obj2)
print(obj1/obj2)
print(obj1//obj2)
print(obj1%obj2)
print(obj1**obj2)
```

```python
# single inheritance
class parent:
    def func1(self):
        print("HEllo parent")
class child(parent):
    def func2(self):
        print("HEllo child")
obj=child()
obj.func1()
obj.func2()
```

```python
# multiple inheritance 
class parent1:
    def func1(self):
        print("hello parent1")
class parent2:
    def func2(self):
        print("hello parent2")
class child(parent1,parent2):
    def func3(self):
        print("Hello child")
obj=child()
obj.func1()
obj.func2()
obj.func3()
```

```python
# multilevel inheritance
class Grandparent:
    def func1(self):
        print("Grandparent")
class parent(Grandparent):
    def func2(self):
        print("parent")
class child(parent):
    def func3(self):
        print("child")
obj=child()
obj.func1()
obj.func2()
obj.func3()
```

```python
# hierarchial inheritance
class parent:
    def func1(self):
        print("Grandparent")
class child1(parent):
    def func2(self):
        print("child1")
class child2(parent):
    def func2(self):
        print("child2")
obj=child1()
obj1=child2()
obj.func1()
obj.func2()
obj1.func1()
obj1.func2()
        
```

```python
# hybrid inheritance
class parent1:
    def func1(self):
        print("Hello parent1")
class parent2:
    def func2(self):
        print("Hello parent2")
class child1(parent1):
    def func3(self):
        print("Hello child1")
class child2(child1,parent2):
    def func4(self):
        print("Hello child2")
obj=child1()
obj.func1()
obj.func3()
obj1=child2()
obj1.func1()
obj1.func2()
obj1.func3()
obj1.func4()
```
