### line styles
'-'       solid line style
'--'      dashed line style
'-.'      dash-dot line style
':'       dotted line style

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([2,4,6,8])
plt.plot(y,'o-.r')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([3,8,1,10])
plt.plot(y,'o-.r',ms=10,mec="blue",mfc="black")  #marker size - ms, mec - marker edge color,mfc - marker face color
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([3,8,1,10])
plt.plot(y,ls="dotted",c="blue",linewidth=20)      #linestyle - ls, color - c , linewidth - lw
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y1=np.array([3,8,1,10])
y2=np.array([6,2,7,11])

plt.plot(y1)      #linestyle - ls, color - c , linewidth - lw
plt.plot(y2)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])
plt.plot(x,y,marker="o")
plt.xlabel("Average pulse",loc="right")
plt.ylabel("Average Burnage",loc="top")
plt.title("Sports Watch Data",loc="left") # loc - location of title  (default - center)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])
font1={'family':'serif','color':'blue','size':20}
font2={'family':'serif','color':'darkred','size':10}
plt.plot(x,y)
plt.xlabel("Average pulse",loc="right",fontdict=font2)
plt.ylabel("Average Burnage",loc="top",fontdict=font2)
plt.title("Sports Watch Data",loc="left",fontdict=font1)
plt.grid(axis='y',color="red")
plt.show()
```

## subplot


```python
import matplotlib.pyplot as plt
import numpy as np
x1=np.array([0,1,2,3])
y1=np.array([3,8,1,10])
plt.subplot(2,1,1)
plt.plot(x1,y1)
plt.xlabel("AAAAAAAA")
plt.ylabel("BBBBBBBB")
plt.title("First")
plt.grid()
x2=np.array([0,1,2,3])
y2=np.array([10,20,30,40])
plt.subplot(2,1,2)
plt.plot(x2,y2)
plt.xlabel("CCCCCCCC")
plt.ylabel("DDDDDDDD")
plt.title("Second")
plt.grid()
plt.suptitle("Hello")
plt.tight_layout()    #to remove overlapping

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x1=np.array([0,1,2,3])
y1=np.array([3,8,1,10])
plt.subplot(1,2,1)
plt.plot(x1,y1)
plt.xlabel("AAAAAAAA")
plt.ylabel("BBBBBBBB")
plt.title("First")
plt.grid()
x2=np.array([0,1,2,3])
y2=np.array([10,20,30,40])
plt.subplot(1,2,2)
plt.plot(x2,y2)
plt.xlabel("CCCCCCCC")
plt.ylabel("DDDDDDDD")
plt.title("Second")
plt.grid()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(10,8))
x1=np.array([0,1,2,3])
y1=np.array([3,8,1,10])
plt.subplot(2,3,1)
plt.plot(x1,y1,color='red',marker="o",ls=":")
plt.xlabel("A")
plt.ylabel("B")
plt.title("First")
plt.grid()

x2=np.array([0,1,2,3])
y2=np.array([10,20,30,40])
plt.subplot(2,3,2)
plt.plot(x2,y2,marker="*",color="blue")
plt.xlabel("C")
plt.ylabel("D")
plt.title("Second",color='yellow')
plt.grid()

x3=np.array([0,1,2,3])
y3=np.array([10,20,30,40])
plt.subplot(2,3,3)
plt.plot(x3,y3)
plt.xlabel("E")
plt.ylabel("F")
plt.title("Third",color='green')
plt.grid()

x4=np.array([0,1,2,3])
y4=np.array([10,20,30,40])
plt.subplot(2,3,4)
plt.plot(x4,y4,color="black",marker="o")
plt.xlabel("G",color="pink")
plt.ylabel("H",color="pink")
plt.title("Fourth")
plt.grid()

x5=np.array([0,1,2,3])
y5=np.array([10,20,30,40])
plt.subplot(2,3,5)
plt.plot(x5,y5,color="orange")
plt.xlabel("I")
plt.ylabel("J")
plt.title("Fifth")
plt.grid()

x6=np.array([0,1,2,3])
y6=np.array([10,20,30,40])
plt.subplot(2,3,6)
plt.plot(x6,y6,color="purple")
plt.xlabel("K")
plt.ylabel("L")
plt.title("Sixth")
plt.grid()
plt.tight_layout()
plt.show()
```

## scatter plot

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([5,7,8,7,2,17,2,9,4,11])
y=np.array([99,86,87,88,111,86,103,87,94,78])
plt.scatter(x,y)
x=np.array([1,3,5,7,9,11,13,15,18])
y=np.array([85,87,90,110,145,98,84,83,95])
# colors=np.array(['red','green','blue','purple','black','brown','orange','pink','yellow'])
colors=np.array([0,10,20,30,40,50,60,70,80])
plt.scatter(x,y,marker="*",c=colors,cmap='turbo',alpha=1)
plt.colorbar()
plt.show()
```
