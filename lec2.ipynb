### special functions in pyhton inheritance

```python
# 1) upper()
class parent:
    def __init__(self):
        self.attr1=50
        self.attr2=60
class child(parent):
    def __init__(self):
        super().__init__()
        self.attr3=45
obj=child()
print(obj.attr3)
print(obj.attr1)
print(obj.attr2)
```

```python
class parent:
    def display(self):
        print("parent")
class child(parent):
    def display(self):
        super().display()
        print("child")
obj=child()
obj.display()
```

```python
# 2) issubclass()
class parent:
    def func1(self):
        print("parent")
class child(parent):
    def func2(self):
        super().display()
        print("child")
print(issubclass(child,parent))
print(issubclass(parent,child))
```

```python
# 3) isinstance()
class parent:
    def func1(self):
        print("parent")
class child(parent):
    def func2(self):
        super().display()
        print("child")
A=child()
B=parent()
print(isinstance(A,child))
print(isinstance(B,child))
print(isinstance(A,parent))
print(isinstance(B,parent))
```

### Method Resolution Order (MRO)
- it denotes a way of programming language resolves a method or attribute.
- MRO defines the order in which the base classes are searched when executing method
- first the method is searched within the class and then it follows the order we specified while inheriting

```python
class A:
    def rk(self):
        print("Class A")
class B(A):
    def rek(self):
        print("Class B")
class C(A):
    def rk(self):
        print("Class C")
r=B()
r.rk()
```

```python
class A:
    def rk(self):
        print("Class A")
class B(A):
    def rk(self):
        print("Class B")
class C(A):
    def rk(self):
        print("Class C")
class D(C,B):
    pass

r=D()
r.rk()
```

```python
class p1:
    def foo(self):
        print("foo from p1")
class p2:
    def foo(self):
        print("foo from p2")
    def bar(self):
        print("bar from p2")
class c1(p1):
    pass
class c2(p2):
    def bar(self):
        print("bar from c2")
class gc(c1,c2):
    pass
obj=gc()
obj.bar()
obj.foo()
```

## Abstract Class
###### syntax
- from abc import ABC
- class class_name(ABC):
- body of the class

```python
from abc import ABC,abstractmethod
class shape(ABC):
    def __init__(self,shape_name):
        self.shape_name=shape_name
    @abstractmethod
    def draw(self):
        pass
class rect(shape):
    def __init__(self):
        super().__init__("rectangle")
    def draw(self):
        print("drawing: ",self.shape_name)
obj=rect()
obj.draw()
```

```python
from abc import ABC,abstractmethod
class shape(ABC):
    def __init__(self,shape_name):
        self.shape_name=shape_name
    @abstractmethod
    def draw(self):
        pass
class rect(shape):
    def __init__(self):
        super().__init__("rectangle")
    def draw_rect(self):
        print("drawing: ",self.shape_name)
    def draw(self):
        pass
obj=rect()
obj.draw_rect()
```

```python
class Book:
    def __init__(self,name,n_author,list_author,publisher,ISBN,year):
        self.name=name
        self.n_author=n_author
        self.list_author=list_author
        self.publisher=publisher
        self.ISBN=ISBN
        self.year=year
    def display(self):
        print("Book name: ",self.name)
        print("No of Author's: ",self.n_author)
        print("list of Author's: ",self.list_author)
        print("publisher: ",self.publisher)
        print("ISBN code: ",self.ISBN)
        print("year: ",self.year)
class courseBook(Book):
    def __init__(self,course_name):
        super().__init__('Ramayan',1,["vyas"],'vyas jii',12364664,1000)
        self.course_name=course_name
    def display(self):
        super().display()
        print("Course name : ",self.course_name)
obj=courseBook("CSE")
obj.display()
```

```python
class student:
    def __init__(self,roll,name,age,totalmarks):
        self.roll=roll
        self.name=name
        self.age=age
        self.totalmarks=totalmarks
    def display(self):
        print("Name : ",self.name)
        print("Age : ",self.age)
        print("roll no : ",self.roll)
        print("Total marks : ",self.totalmarks)
    def __eq__(self,other):
        if self.totalmarks==other.totalmarks:
            return True
        else:
            return False
obj1=student(21,'yatrik',21,25)
obj2=student(22,'meet',20,25)
if obj1==obj2:
    print("Equal")
else:
    print("Not Equal")
```
